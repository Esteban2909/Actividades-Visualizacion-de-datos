## Visualización 1 : 

### Descripción: 

La visualización representa un estudio sobre las lenguas mas utilizadas en el mundo.






### 1. What ? 

De la visualización se puede inferir que los datos pueden estar agrupados en una tabla, discriminados en varias variables tanto categoricas como cuantitativas, de las cuales se puede destacar dentro de las categoricas el ***pais*** y el ***idioma***, y respecto a las cuantitativas la ***población*** por pais que habla determinada lengua.


### 2. Why ?

Para identificar el **why** se puede empezar analizando sus componentes. En primer lugar se tienen las acciones de alto nivel, donde el **Discover** juega un papel relevante, ya que uno de los objetivos principales de esta visualización puede ser descubrir información relevante de los datos, otra acción de alto nivel que se puede identificar es el **present**, donde por medio de una infografia se presenta la visualización. 

Respecto a las acciones de bajo y medio nivel se puede resaltar la parte del **Query**, ya que con la visualización se puede identificar datos atipicos o relevantes o se puede realizar comparaciones entre categorias de las variables.

En cuanto a los objetivos principales se puede observar que la visualización tiene el fin de mostrar correlaciones o dependencias entre algunas variables, por ejemplo el area geografica con la cantidad de población y por ende la cantidad de personas que hablan cierta lengua.


### 3. How ?

Se puede evidenciar que en la visualización se utiliza una marca de **Area** para distinguir las proporciones de personas y paises que hablan cierto idioma, ademas se utiliza el canal de **color** para distinguir la región geografica a la que pertenece cada pais, en este caso delimitada por continentes.


### Observaciones sobre la visualización:

En mi opinión la visualización es bastante clara en cuanto al **What** y el **why**, en cuanto al **How** se siguen correctamente las reglas principales sobre el color para distinguir de categorias, pero la forma de algunas areas puede generar confución a primera vista, ya que el tamaño de las figuras dentro del circulo representa la cantidad de personas que hablan cierto idioma, es decir entre mayor sea el tamaño del area mayor es la cantidad de personas que habla cierto idioma. En algunas casos  el tamaño entre figura y figura a primera vista no varia demaciado pero en la cantidad de personas si se puede observar una gran diferencia. 


## Visualización 2:

### Descripción:

Esta visualización es la representación del comportamiento de Covid 19 en el pais de Chile.

#### Link de la visualización:

[Enlace Visualizacion](https://covid.visorterritorial.cl/)


![Visualizacion](C:/Users/User/Desktop/Visualización-2.jpeg)


### 1. What ?

De esta visualización se puede decir que el Dataset es una geometria, ya que hay información representada en forma de mapa, donde se pueden encontrar variables categoricas como las ciudades o regiones de Chile y una variable cuantitativa como lo es el numero de casos activos de covid_19 en el pais. Se puede decir que la visualización tambien presenta una variable secuencial como el tiempo, ya que los datos son actualizados dia a dia.


### 2. Why ? 

Al igual que en la visualización anterior para la identificación del **Why** se deben analizar las acciones de alto nivel. En este caso tambien se debe tener en cuenta el **Discover** ya que una de las finalidades de esta visualización es descubrir aspectos relevantes sobre el comportamiento del covid-19 Chile.


En cuanto a las acciones de bajo y medio nivel se puede resaltar el **Query** sobre todo en la parte de **identificar** y **comparar** ya que se observar los sectores del pais con mas o menos casos activos de covid-19 y hacer comparaciones entre ellos.

En cuanto a los objetivos de la visualización puede ser identificar atributos, como la dependencia o correlación entre las variables. Por ejemplo si dependiendo la zona del pais se presentan mas o menos casos de covid_19

### 3. How ?

La visualización cuenta con marco de area para delimitar los sectores o regiones del pais, de posición horizontal y vertical para determinar cordenadas dentro del mapa y cuenta con un canal de color para los atributos cuantitativos ordenados como el numero de casos activos por región.

### Observaciones sobre la visualización:

Esta visualización esta muy bien realizada, los marcos y canales son los indicados para la presentación de la información que se tiene, ademas cumple con una regla importante en las visualizaciones como lo es el uso de un mismo color con distintas tonalidades para variables cuantitativas ordenadas y se puede evidenciar facilmente los aspectos del **What** y el **Why**
