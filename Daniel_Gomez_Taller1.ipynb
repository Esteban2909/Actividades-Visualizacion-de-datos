## Visualizacin 1 : 

### Descripcin: 

La visualizacin representa un estudio sobre las lenguas mas utilizadas en el mundo.






### 1. What ? 

De la visualizacin se puede inferir que los datos pueden estar agrupados en una tabla, discriminados en varias variables tanto categoricas como cuantitativas, de las cuales se puede destacar dentro de las categoricas el ***pais*** y el ***idioma***, y respecto a las cuantitativas la ***poblacin*** por pais que habla determinada lengua.


### 2. Why ?

Para identificar el **why** se puede empezar analizando sus componentes. En primer lugar se tienen las acciones de alto nivel, donde el **Discover** juega un papel relevante, ya que uno de los objetivos principales de esta visualizacin puede ser descubrir informacin relevante de los datos, otra accin de alto nivel que se puede identificar es el **present**, donde por medio de una infografia se presenta la visualizacin. 

Respecto a las acciones de bajo y medio nivel se puede resaltar la parte del **Query**, ya que con la visualizacin se puede identificar datos atipicos o relevantes o se puede realizar comparaciones entre categorias de las variables.

En cuanto a los objetivos principales se puede observar que la visualizacin tiene el fin de mostrar correlaciones o dependencias entre algunas variables, por ejemplo el area geografica con la cantidad de poblacin y por ende la cantidad de personas que hablan cierta lengua.


### 3. How ?

Se puede evidenciar que en la visualizacin se utiliza una marca de **Area** para distinguir las proporciones de personas y paises que hablan cierto idioma, ademas se utiliza el canal de **color** para distinguir la regin geografica a la que pertenece cada pais, en este caso delimitada por continentes.


### Observaciones sobre la visualizacin:

En mi opinin la visualizacin es bastante clara en cuanto al **What** y el **why**, en cuanto al **How** se siguen correctamente las reglas principales sobre el color para distinguir de categorias, pero la forma de algunas areas puede generar confucin a primera vista, ya que el tamao de las figuras dentro del circulo representa la cantidad de personas que hablan cierto idioma, es decir entre mayor sea el tamao del area mayor es la cantidad de personas que habla cierto idioma. En algunas casos  el tamao entre figura y figura a primera vista no varia demaciado pero en la cantidad de personas si se puede observar una gran diferencia. 


## Visualizacin 2:

### Descripcin:

Esta visualizacin es la representacin del comportamiento de Covid 19 en el pais de Chile.

#### Link de la visualizacin:

[Enlace Visualizacion](https://covid.visorterritorial.cl/)


![Visualizacion](C:/Users/User/Desktop/Visualizacin-2.jpeg)


### 1. What ?

De esta visualizacin se puede decir que el Dataset es una geometria, ya que hay informacin representada en forma de mapa, donde se pueden encontrar variables categoricas como las ciudades o regiones de Chile y una variable cuantitativa como lo es el numero de casos activos de covid_19 en el pais. Se puede decir que la visualizacin tambien presenta una variable secuencial como el tiempo, ya que los datos son actualizados dia a dia.


### 2. Why ? 

Al igual que en la visualizacin anterior para la identificacin del **Why** se deben analizar las acciones de alto nivel. En este caso tambien se debe tener en cuenta el **Discover** ya que una de las finalidades de esta visualizacin es descubrir aspectos relevantes sobre el comportamiento del covid-19 Chile.


En cuanto a las acciones de bajo y medio nivel se puede resaltar el **Query** sobre todo en la parte de **identificar** y **comparar** ya que se observar los sectores del pais con mas o menos casos activos de covid-19 y hacer comparaciones entre ellos.

En cuanto a los objetivos de la visualizacin puede ser identificar atributos, como la dependencia o correlacin entre las variables. Por ejemplo si dependiendo la zona del pais se presentan mas o menos casos de covid_19

### 3. How ?

La visualizacin cuenta con marco de area para delimitar los sectores o regiones del pais, de posicin horizontal y vertical para determinar cordenadas dentro del mapa y cuenta con un canal de color para los atributos cuantitativos ordenados como el numero de casos activos por regin.

### Observaciones sobre la visualizacin:

Esta visualizacin esta muy bien realizada, los marcos y canales son los indicados para la presentacin de la informacin que se tiene, ademas cumple con una regla importante en las visualizaciones como lo es el uso de un mismo color con distintas tonalidades para variables cuantitativas ordenadas y se puede evidenciar facilmente los aspectos del **What** y el **Why**
